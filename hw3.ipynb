## Самое популярное мужское и самое популярное женское имя на корабле

```python
def title(name):
    return name.split()[1][:-1]
def first_name(name):
    if title(name) == 'Mrs': 
        try: 
            first_name = name.split('(')[1].split()[0]
            return first_name
        except:
            return name.split()[2]
    else: return name.split()[2]
```

```python
df['First name'] = df['Name'].apply(lambda x: first_name(x))
```

```python
name = df[df['Sex'] == 'female']['First name'].value_counts().index[0] # самое популярное женское имя
print(f'Самое популярное женское имя на корабле {name}.')
```

```python
name = df[df['Sex'] == 'male']['First name'].value_counts().index[0] # самое популярное мужское имя
print(f'Самое популярное мужское имя на корабле {name}.')
```

## Самое популярное мужское и самое популярное женское имя на корабле в каждом классе

```python
for sex in ['female', 'male']:
    if sex == 'female': sex_ru = 'женское'
    else: sex_ru = 'мужское'
    for pclass in range(3):
        name = df[df['Sex'] == sex].groupby('Pclass')['First name'].value_counts().index[pclass][1]
        print(f'Самое популярное {sex_ru} имя в {pclass + 1} классе {name}.')
```
